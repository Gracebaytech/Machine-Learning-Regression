# Model Development

This section includes below steps:

Data preparation in order to feed the model

Develop different models and compare them.

```python
# importing python libaries

import time
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVR
from  sklearn.linear_model import BayesianRidge
from catboost import CatBoostRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.linear_model import LinearRegression
from xgboost.sklearn import XGBRegressor
from lightgbm import  LGBMRegressor


from sklearn.metrics import mean_absolute_error, mean_squared_error,mean_absolute_error,r2_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# splitting dataset in train abf val
X=train.iloc[:,0: -1]
Y = train.iloc[:, -1]
test_size=0.20
seed=42
X_train, X_Val, Y_train, Y_Val = train_test_split(X, Y, test_size=test_size,random_state=seed)
```

```python
def Models(model, name, d):
    print("Working on {} model".format(name))
    
    
    cla = model
    cla.fit(X_train, Y_train)
    
   
    
    predicted = cla.predict(X_Val)
   
    
    MAE=mean_absolute_error(Y_Val,predicted)
    MSE=mean_squared_error(Y_Val,predicted)
    RMSE=mean_squared_error(Y_Val,predicted,squared=False)
    MAEA=mean_absolute_error(Y_Val,predicted)
    R2=r2_score(Y_Val,predicted)
    
    
    
    d['Name'].append(name)
    #d['Training ACU'].append(tr_auc)
    #d['Testing ACU'].append(te_auc)
    d["RMSE"].append(RMSE)
    d['MAE'].append(MAE)
    d['MSE'].append(MSE)
    d['MAEA'].append(MAEA)
    d['R2'].append(R2)
    
    
    
    
    print("**********"*5)
    print()
    return d
    
    
    
    
     
```

```python

d = {'Name' : [], 'RMSE': [], 'R2': [], 'MAE': [], 'MSE': [],'MAEA': []}


models = [ [GradientBoostingRegressor(loss="squared_error",learning_rate=0.2,n_estimators=1000), 'GradientBoostRegressor'],
          [ElasticNet(max_iter=2000,normalize=True), 'ElasticNet'], 
          [LGBMRegressor(), 'LightGBM'],
          [SGDRegressor(l1_ratio=0.5,), "SGDRegressor"],

          [CatBoostRegressor(), 'CatBoost'], 
          [KernelRidge(alpha=1.0), 'kernal Ridge'],
          [LinearRegression(), 'Linear Regression'], 
          [SVR(C=1.0,epsilon=0.2), 'Support Vector Ratio' ],
         [XGBRegressor(objective="reg:linear"),"XGBRegressor"],
         [BayesianRidge(),"BayesianRidge"]
         ]




for model in models:
    d = Models(model[0], model[1], d)

acu_data = pd.DataFrame(data = d)
```

```python
acu_data
```

```python
# Hyperparameter Optimization
xgb_parms={"max_depth":[3,4,5,6,10,12,15],
          "learning_rate":[0.05,0.10,0.15,0.20,0.25,0.30],
          #"min_samples_split":[1,3,5,7],
          "gamma":[0.0,0.1,0.2,0.3,0.4],
          "colsample_bytree":[0.3,0.4,0.5,0.6]}


rf_parms={"n_estimators": np.arange(100,1500,100),
         "max_depth": np.arange(1,20),
          
         "max_features":['auto','sqrt','log2']}
```

```python
from sklearn.tree import DecisionTreeRegressor
from  sklearn.ensemble import RandomForestRegressor
```

```python
# ~~~~~~~~~~~~~~~~~~~~~~~~~
# Decision Tree sklearn
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~

model_11 = DecisionTreeRegressor()

model_11.fit(X_train, Y_train)

model_11_pred = model_11.predict(X_Val)

mean_abs_error_m11 = mean_absolute_error(model_11_pred, Y_Val)

print("Mean Absolute Error of Model 11 : ", mean_abs_error_m11)
```

```python
# ~~~~~~~~~~~~~~~~~~~~~~~~~
# Random Forest sklearn
# ~~~~~~~~~~~~~~~~~~~~~~~~~

model_12 = RandomForestRegressor()

model_12.fit(X_train, Y_train.values.ravel())

model_12_pred = model_12.predict(X_Val)

mean_abs_error_m12 = mean_absolute_error(model_12_pred, Y_Val)

print("Mean Absolute Error of Model 12 : ", mean_abs_error_m12)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search_rf = RandomizedSearchCV(model_12, param_distributions = rf_parms, n_iter = 5, scoring = 'roc_auc', n_jobs=-1, 
                                  cv = 5, verbose = 3)
```

```python
# Hyperparameter Optimization
start_time = time.time()

random_search_rf.fit(X_train, Y_train.values.ravel())

end_time = time.time()

print("Time taken : {0}".format(end_time - start_time))
```

```python
random_search_rf.best_estimator_
```

```python
# ~~~~~~~~~~~~~~~~~~~~~~~~~
# XGB Regressor sklearn
# ~~~~~~~~~~~~~~~~~~~~~~~~~

model_8 = XGBRegressor()

model_8.fit(X_train, Y_train)

model_8_pred = model_8.predict(X_Val)

mean_abs_error_m8 = mean_absolute_error(model_8_pred, Y_Val)

print("Mean Absolute Error of Model 8 : ", mean_abs_error_m8)
```

```python
random_search = RandomizedSearchCV(model_8, param_distributions = xgb_parms, n_iter = 5, n_jobs=-1, 
                                   cv = 5, verbose = 3)
```

```python
 #treatment of the outlier
from scipy import stats
import numpy as np
train1=train.copy()
print("shape of data before outliers :",train1.shape)
# using z score to remove the outier
z=np.abs(stats.zscore(train))
W_train=train[(z<3).all(axis=1)]
print("shape of data after outlier:",W_train.shape)


W_train.reset_index(inplace=True)
```

```python
X=W_train.iloc[:,0: -1]
Y =W_train.iloc[:, -1]
test_size=0.20
seed=42
X_Wtrain, X_Wval, Y_Wtrain, Y_Wval = train_test_split(X, Y, test_size=test_size,random_state=seed)
```

```python
def Models(model, name, d):
    print("Working on {} model".format(name))
    
    
    cla = model
    cla.fit(X_Wtrain, Y_Wtrain)
    
   
    
    predW = cla.predict(X_Wval)
   
    
    MAE=mean_absolute_error(Y_Wval,predW)
    MSE=mean_squared_error(Y_Wval,predW)
    RMSE=mean_squared_error(Y_Wval,predW,squared=False)
    MAEA=mean_absolute_error(Y_Wval,predW)
    R2=r2_score(Y_Wval,predW)
    
    
    
    d['Name'].append(name)
    #d['Training ACU'].append(tr_auc)
    #d['Testing ACU'].append(te_auc)
    d["RMSE"].append(RMSE)
    d['MAE'].append(MAE)
    d['MSE'].append(MSE)
    d['MAEA'].append(MAEA)
    d['R2'].append(R2)
    
    
    
    
    print("**********"*5)
    print()
    return d
    
    
    
    
     
```
